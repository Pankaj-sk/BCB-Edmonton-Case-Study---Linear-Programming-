```python
from pulp import *
import pandas as pd
import numpy as np
```

```python
n_agents = 5
n_branches = 35
```

```python
df = pd.read_csv('cost.csv')
# print(df.head())


cost_matrix = df.to_numpy()
print(cost_matrix)
```

```python
df = pd.read_csv('cust_demands.csv')
# print(df.head())

cust_demands = df.to_numpy()
print(cust_demands)

# Demand Matrix
#cust_demands = np.array([35000, 22000, 18000, 30000])
```

```python
# Supply Matrix
warehouse_supply = np.array([8160,8160,8160,8160,8160])
print(warehouse_supply)
```

```python
model = LpProblem("Supply-Demand-Problem", LpMinimize)
```

```python
variable_names = [str(i)+"_"+str(j) for j in range(1, n_branches +1) for i in range(1, n_agents +1)]


print("Variable Indices:", variable_names)
```

```python
DV_variables = LpVariable.matrix("X", variable_names, cat = "Integer", lowBound= 0 )

allocation = np.array(DV_variables).reshape(35,5)

allocation = np.transpose(allocation)

print("Decision Variable/Allocation Matrix: ")

print(allocation)
```

```python
obj_func = lpSum(allocation*cost_matrix)
print(obj_func)
model +=  obj_func
print(model)
```

```python
#Supply Constraints

for i in range(n_agents):
    print(lpSum(allocation[i][j] for j in range(n_branches)) <= warehouse_supply[i])
    model += lpSum(allocation[i][j] for j in range(n_branches)) <= warehouse_supply[i] , "Supply Constraints " + str(i)

```

```python
# Demand Constraints
for j in range(n_branches):
    
    print(lpSum(allocation[i][j] for i in range(n_agents)) >= cust_demands[0][j])
    
    model += lpSum(allocation[i][j] for i in range(n_agents)) >= cust_demands[0][j] , "Demand Constraints " + str(j)
    
    
    
```

range(n_branches)

```python
model.writeLP(filename='temp.lp')
```

```python
#model.solve()
model.solve(PULP_CBC_CMD())

status =  LpStatus[model.status]

print(status)
```
